```python
import pandas as pd
import os

def load_real_data():
    base_path = os.path.dirname(os.path.dirname(__file__))
    file_path = os.path.join(base_path, 'data', 'raw', 'carbon_nanotubes.csv')
    
    df = pd.read_csv(file_path, sep=';')
    
    if df.shape[1] <= 1:
        df = pd.read_csv(file_path, sep=',')

    df.columns = [c.strip().replace('"', '').replace("'", "").lower() for c in df.columns]
    
    for col in df.columns:
        if 'chiral_indice_n' in col: df.rename(columns={col: 'chiral_n'}, inplace=True)
        if 'chiral_indice_m' in col: df.rename(columns={col: 'chiral_m'}, inplace=True)
        if 'coordinate_u' in col: df.rename(columns={col: 'u_coord'}, inplace=True)
        if 'coordinate_v' in col: df.rename(columns={col: 'v_coord'}, inplace=True)
        if 'coordinate_w' in col: df.rename(columns={col: 'target_w'}, inplace=True)

    return df
```
